```r
data_by_state<-NULL
for (i in 1:length(uni)){
    data_by_state[i] <- length(filter(data, State==uni[i])$ID) 
    }
```

```r
barplot_data<-cbind(uni,data_by_state)
```

```r
barplot_data_frame<-data.frame(barplot_data)
barplot_data_frame
```

```r
set_plot_dimensions <- function(width_choice, height_choice) {
        options(repr.plot.width=width_choice, repr.plot.height=height_choice)
        }
```

```r
set_plot_dimensions(50,25)
par(bg = "white")
x<-reorder(uni,-data_by_state)
y<-data_by_state

ggplot(barplot_data_frame,aes(x=x, y=y,fill=y))+
    geom_bar(stat = 'identity')+
    scale_fill_gradient(low="blue", high="red")+
    theme(axis.text.x = element_text(colour = "red", size = 40,angle=90),
        axis.text.y = element_text(colour = "blue", size = 40))
```
